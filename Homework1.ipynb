## Problem 1: Root-finding Algorithms

### Part 1: Newton-Raphson Method

```python
def f(x):
    return x * np.sin(3*x) - np.exp(x)

def dfdx(x):
    return np.sin(3*x) + 3*x*np.cos(3*x) - np.exp(x)

# Newton-Raphson
x = [-1.6]
while np.abs(f(x[-1])) > 1e-6:
    x.append(x[-1] - f(x[-1]) / dfdx(x[-1]))
x = np.array(x)
niter_NR = len(x)

#######
# TODO: Define the vector of x-values found using the Newton-Raphson method.
# A1 = ??? (numpy.ndarray)
#######
A1 = x
```

```python
# Plot f(x) and search path
plt.figure(figsize=(12,8))
domain = np.linspace(np.min(x), np.max(x), 1001)
plt.plot(domain, f(domain), 'k-')
plt.plot(A1, f(A1), 'b.-', linewidth=0.5)
plt.title('Newton-Raphson Search Path')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)

# Plot convergence history of f(x)
plt.figure()
plt.semilogy(np.abs(f(A1)), 'k.-')
plt.title('Newton-Raphson Convergence')
plt.xlabel('$n$')
plt.ylabel('$|f(x_n)|$')
plt.grid(True)
plt.axhline(1e-6, c='r', linewidth=0.5, label='convergence criterion')
plt.legend()

# Plot convergence history of x
plt.figure()
plt.plot(A1, 'k.-')
plt.title("Newton-Raphson Results")
plt.ylabel("$x_n$")
plt.xlabel("$n$")
plt.grid(True)

plt.show()
```

### Part 2: Bisection Method

```python
bracket = np.array([-0.7, -0.4])
x = [np.mean(bracket)]
while np.abs(f(x[-1])) > 1e-6:
    if f(x[-1]) > 0:
        bracket[0] = x[-1]
    else:
        bracket[1] = x[-1]
    x.append(np.mean(bracket))
x = np.array(x)
niter_bisect = len(x)

#######
# TODO: Define the vector of x_mid values found using the Bisection method.
# A2 = ??? (numpy.ndarray)
#######
A2 = x
```

```python
# Plot f(x) and search path
plt.figure(figsize=(12,8))
domain = np.linspace(np.min(x), np.max(x), 1001)
plt.plot(domain, f(domain), 'k-')
plt.plot(A2, f(A2), 'b.-', linewidth=0.5)
plt.title('Bisection Search Path')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)

# Plot convergence history of f(x)
plt.figure()
plt.semilogy(np.abs(f(A2)), 'k.-')
plt.title('Bisection Convergence')
plt.xlabel('$n$')
plt.ylabel('$|f(x_n)|$')
plt.grid(True)
plt.axhline(1e-6, c='r', linewidth=0.5, label='convergence criterion')
plt.legend()

# Plot convergence history of x
plt.figure()
plt.plot(A2, 'k.-')
plt.title("Bisection Results")
plt.ylabel("$x_n$")
plt.xlabel("$n$")
plt.grid(True)

plt.show()
```
